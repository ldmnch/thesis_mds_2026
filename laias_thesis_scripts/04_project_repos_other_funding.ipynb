```python
from ossdirectory import fetch_data
from ossdirectory.fetch import OSSDirectory
from typing import List
```

```python
import pandas as pd
```

```python
OSSDirectory = fetch_data()
projects: List[dict] = OSSDirectory.projects;
collections: List[dict] = OSSDirectory.collections;
```

```python
# extract
projects_data = pd.DataFrame(projects)
```

```python
projects_data['repository_url'] = projects_data['github'].str[0].str.get('url')
```

```python
projects_data = projects_data[['name', 'description', 'repository_url']]
```

```python
projects_data
```

```python
funding_data = pd.read_csv("https://github.com/opensource-observer/oss-funding/blob/main/data/funding_data.csv?raw=true")
```

```python
funding_data_final = funding_data.merge(projects_data, left_on='to_project_name', right_on='name', how='left')
```

```python
funding_data_final.dropna(subset=['repository_url'], inplace=True)
```

```python
funding_data_final.to_csv("../data/proc/project_repos_other_funding.csv", index=False)
```
